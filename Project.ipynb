#Tratamiento de datos nulos

En este paso identifique los datos nulos para poder definir que ibamos a hacer con ellos. Al tratarse de un ecommerce no deberiamos borrar ninguna transacción ya que perderiamos información del comportamiento del usuario, por ende decidí reemplazar con ceros las columnas que tenian nulos exceptuando la columna "Categoría" la cual se reemplazo por el valor "Otros".


--------
# Handling of null/missing values
In this step, I identified the null/missing data in order to define what to do with them. Since it's an e-commerce, we shouldn't delete any transactions as it would result in losing information on user behavior. Therefore, I decided to replace the null/missing values in columns with zeros, except for the "Category" column, which was replaced with the value "Others".

```python
df3.isnull().sum()
```

```python
#Definimos las columnas que queremos reemplazar con cero
columnas_para_reemplazar_con_cero= df3[["Fecha Entrega","ID Empresa","Empresa","Apellido","Calle","Nro","Piso","Depto","Localidad","Barrio","Tel","Lista Blanca/ Dominio",
          "Tracking","DetalleMedioDePago","factura","Ref. Interna","Ref. Externa","Porcentaje de Descuentos"]]
      
```

```python
#Reemplazamos con "Otros" a las categorias que tienen valores nulos ya que no esta definida la categoria
df3['Categoria'].fillna("Otros", inplace=True)
```

```python
#Corremos un bucle para que se reemplace cada valor nulo de las columnas definidas con 0
for col in columnas_para_reemplazar_con_cero:
    df3[col].fillna(0, inplace=True)
```

```python

# agrupamos los datos por categoría y sitio y calculamos  la suma total de "total" para cada grupo
total_sales = df3.groupby(["Categoria", "Sitio"])["Total"].sum().reset_index()

# creamos un gráfico de barras en Seaborn para visualizar las ventas totales por categoría y sitio
plt.figure(figsize=(12, 6))
sns.barplot(x=df3["Categoria"], y=df3["Total"], hue=df3["Sitio"], data=total_sales)
plt.title("Ventas totales por categoría y sitio")
plt.xlabel("Categoría")
plt.ylabel("Ventas totales")
plt.show()

# creamos un gráfico de línea en Seaborn para visualizar la evolución de las ventas totales por mes
monthly_sales = ventas_aprobadas.groupby("Mes")["Total"].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x=df3["Mes"], y=df3["Total"], data=monthly_sales)
plt.title("Evolución de las ventas totales por mes")
plt.xlabel("Mes")
plt.ylabel("Ventas totales")
plt.show()

# creamos un gráfico de dispersión en Seaborn para visualizar la relación entre el precio total y la categoría de los productos vendidos
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df3["Total"], y=df3["Categoria"], data=df3)
plt.title("Relación entre precio total y categoría de productos vendidos")
plt.xlabel("Precio total")
plt.ylabel("Categoría")
plt.show()

# creamos un gráfico de barras en Matplotlib para visualizar las ventas totales por categoría y sitio
plt.figure(figsize=(12, 6))
plt.bar(total_sales["Categoria"] + "-" + total_sales["Sitio"], total_sales["Total"])
plt.title("Ventas totales por categoría y sitio")
plt.xlabel("Categoría - Sitio")
plt.ylabel("Ventas totales")
plt.xticks(rotation=90)
plt.show()

# creamos un gráfico de línea en Matplotlib para visualizar la evolución de las ventas totales por mes 
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales["Mes"], monthly_sales["Total"])
plt.title("Evolución de las ventas totales por mes")
plt.xlabel("Mes")
plt.ylabel("Ventas totales")
plt.show()

# creamos un gráfico de dispersión en Matplotlib para visualizar la relación entre el precio total y la categoría de los productos vendidos 
plt.figure(figsize=(12, 6))
plt.scatter(df3["Total"], df3["Categoria"])
plt.title("Relación entre precio total y categoría de productos vendidos")
plt.xlabel("Precio total")
plt.ylabel("Categoría")
plt.show()
```
